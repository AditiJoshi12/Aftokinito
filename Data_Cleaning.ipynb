# Data Cleaning

```python
(auto.isnull().sum())
```

We shall drop "Engine_Type", "USB_Ports", "Battery", "Electric_Range" since more than 1200 cells in those fields are empty. 

```python
auto.drop(["Engine_Type", "USB_Ports", "Battery", "Electric_Range"], axis=1, inplace=True)
```

```python
def return_sorted_cols(nullcolumns):
    empty_cols = {}
    n = len(nullcolumns)
    # empty_cols contains number of null cells in a given column
    for i in range(n):
        empty_cols[cols[i]] = nullcolumns[i]
    
    keys = list(empty_cols.keys())
    
    # If the number of null values in a column is more than 1000, then the column cannot give us sufficient information.
    # Hence, we shall drop those.
    for key in keys:
        if empty_cols[key] >= 1000:
            empty_cols.pop(key)
    
    # Sort the dictionary according to numbre of null cells and print
    for col in sorted(empty_cols, key=empty_cols.get, reverse=False):
        print(col)
        print(empty_cols[col])
        pprint(auto[col].unique())
        print('\n')
```

```python
nullcolumns = list(auto.isnull().sum())
return_sorted_cols(nullcolumns)
```

For fields like Central_Locking, Power_Seats, EBA, Bluetooth, Parking_Assistance, Keyless_Entry, etc. it is more likely that the "nan"/ empty fields imply that these features/services are not available for the given model. Hence, we replace the "nan" cells with "No".

```python
fields_to_be_replaced_by_no = ['Power_Seats', 'EBA_(Electronic_Brake_Assist)', 'Bluetooth', 'Parking_Assistance', 
                              'Keyless_Entry', 'Distance_to_Empty', 'EBD_(Electronic_Brake-force_Distribution)', 
                              'Fasten_Seat_Belt_Warning', 'Central_Locking', 'Door_Ajar_Warning', 
                              'ABS_(Anti-lock_Braking_System)', 'Power_Windows', 'Child_Safety_Locks', 
                              'Engine_Immobilizer', 'Power_Steering', 'Ventilation_System']
```

```python
for field in fields_to_be_replaced_by_no:
    auto[field].replace(np.nan, "No", inplace=True)
```
