Importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from termcolor import colored
import warnings
warnings.filterwarnings('ignore')
from pprint import pprint
```

```python
auto_num = pd.read_csv("datasets/clean_dataset.csv", index_col=0)
auto_num.head()
```

```python
cols = list(auto_num.columns)
```

```python
dtypes = list(auto_num.dtypes)
n = len(dtypes)
for i in range(n):
    if dtypes[i] == 'object':
        print(cols[i])
        print(auto_num[cols[i]].unique())
        print("\n")
```

```python
df = auto_num.copy()
```

```python
df
```

```python
yes_no = ['EBA_(Electronic_Brake_Assist)', 'Fasten_Seat_Belt_Warning', 'EBD_(Electronic_Brake-force_Distribution)', 
          'Door_Ajar_Warning', 'ABS_(Anti-lock_Braking_System)', 'Engine_Immobilizer', 'Distance_to_Empty', 
          'Child_Safety_Locks', 'Central_Locking', 'Bluetooth']
```

```python
for field in yes_no:
    df[field].replace("Yes", 1, inplace=True)
    df[field].replace("No", 0, inplace=True)
```

```python
df["Fuel_System"].replace("Injection", 0, inplace=True)
df["Fuel_System"].replace("PGM - Fi", 1, inplace=True)
```

```python
df['Handbrake'].replace('Manual', 0, inplace=True)
df['Handbrake'].replace('Automatic', 1, inplace=True)
```

```python
df["Emission_Norm"].replace('BS III', 3, inplace=True)
df["Emission_Norm"].replace('BS IV', 4, inplace=True)
df["Emission_Norm"].replace('BS VI', 6, inplace=True)
```

```python
df["Power_Windows"].replace('No', 1, inplace=True)
df["Power_Windows"].replace('Only Front Windows', 2, inplace=True)
df["Power_Windows"].replace('All Windows', 3, inplace=True)
```

```python
df['Power_Seats'].replace('No', 0, inplace=True)
df['Power_Seats'].replace(['Yes', 'Power seats'], 1, inplace=True)
df['Power_Seats'].replace('Yes, with memory', 2, inplace=True)
```

```python
'''
Keyless_Entry(Remote, Smart Key = Smart Key, Remote > Smart Key > Remote > Yes > No)
'''

df["Keyless_Entry"].replace('No', 0, inplace=True)
df["Keyless_Entry"].replace('Yes', 1, inplace=True)
df["Keyless_Entry"].replace('Remote', 2, inplace=True)
df["Keyless_Entry"].replace('Smart Key', 3, inplace=True)
df["Keyless_Entry"].replace(['Smart Key, Remote', 'Remote, Smart Key'], 4, inplace=True)
```

```python
'''
Seats_Material(Leather>Vinyl>Polyurethene>Fabric)
'''

df["Seats_Material"].replace('Fabric', 1, inplace=True)
df["Seats_Material"].replace('Polyurethene', 2, inplace=True)
df["Seats_Material"].replace('Vinyl', 3, inplace=True)
df["Seats_Material"].replace('Leather', 4, inplace=True)
```

```python
'''
Type(DCT>CVTAMT> Automatic> Manual)
'''

df["Type"].replace('Manual', 1, inplace=True)
df["Type"].replace('Automatic', 2, inplace=True)
df["Type"].replace('AMT', 3, inplace=True)
df["Type"].replace('CVT', 4, inplace=True)
df["Type"].replace('DCT', 5, inplace=True)
```

```python
'''
Seat_Height_Adjustement
(Electric Adjustment with Memory > Electric Adjustment > Semi Automatic Adjustment > Manual Adjustment)
'''


df["Seat_Height_Adjustment"].replace('Manual Adjustment', 1, inplace=True)
df["Seat_Height_Adjustment"].replace('Semi Automatic Adjustment', 2, inplace=True)
df["Seat_Height_Adjustment"].replace('Electric Adjustment', 3, inplace=True)
df["Seat_Height_Adjustment"].replace('Electric Adjustment with Memory', 4, inplace=True)
```

```python
'''
Parking_Assistance

'No' < 'Yes' < 'Rear sensors' < 'Rear sensors, Rear sensors with camera' = 'Rear sensors with camera, Rear sensors'
= 'Rear sensors with camera' < 'Front sensors, Rear sensors' =  'Rear sensors, Front sensors' <
'Front and rear sensors with camera' < 'Front & rear sensors with 360 degree view' 
 < 'Front and rear sensors with camera, Front & rear sensors with 360 degree view'
 
'''

df['Parking_Assistance'].replace('No', 0, inplace=True)
df['Parking_Assistance'].replace('Yes', 1, inplace=True)
df['Parking_Assistance'].replace('Rear sensors', 2, inplace=True)
df['Parking_Assistance'].replace(['Rear sensors, Rear sensors with camera','Rear sensors with camera, Rear sensors', 
                                  'Rear sensors with camera'], 3, inplace=True)
df['Parking_Assistance'].replace(['Front sensors, Rear sensors', 'Rear sensors, Front sensors'], 4, inplace=True)
df['Parking_Assistance'].replace('Front and rear sensors with camera', 5, inplace=True)
df['Parking_Assistance'].replace('Front & rear sensors with 360 degree view', 6, inplace=True)
df['Parking_Assistance'].replace('Front and rear sensors with camera, Front & rear sensors with 360 degree view', 
                                 6, inplace=True)
```

```python
'''

Ventilation_System

'No' < 'Yes' < 'Heater' < 'Air Conditioning with cooling only' < 'Manual Air conditioning with cooling and heating' =
'Heater, Manual Air conditioning with cooling and heating' < 
'Manual Air conditioning with cooling and heating, Fully automatic climate control' = 'Fully automatic climate control' =
'Heater, Fully automatic climate control' <  '2 Zone Climate Control' <  
'Fully automatic climate control, 2 Zone Climate Control' =  '2 Zone Climate Control, Fully automatic climate control' <
'3 Zone climate control' <  'Fully automatic climate control, 3 Zone climate control' < '4 Zone climate control' < 
'Fully automatic climate control, 4 Zone climate control'

'''

df['Ventilation_System'].replace('No', 0, inplace=True)
df['Ventilation_System'].replace('Yes', 1, inplace=True)
df['Ventilation_System'].replace('Heater', 2, inplace=True)
df['Ventilation_System'].replace('Air Conditioning with cooling only', 3, inplace=True)
df['Ventilation_System'].replace('Manual Air conditioning with cooling and heating', 4, inplace=True)
df['Ventilation_System'].replace('Heater, Manual Air conditioning with cooling and heating', 4, inplace=True)
df['Ventilation_System'].replace(['Manual Air conditioning with cooling and heating, Fully automatic climate control', 
                                  'Fully automatic climate control', 'Heater, Fully automatic climate control'],
                                 5, inplace=True)
df['Ventilation_System'].replace('2 Zone Climate Control', 6, inplace=True)
df['Ventilation_System'].replace(['Fully automatic climate control, 2 Zone Climate Control',
                                  '2 Zone Climate Control, Fully automatic climate control'], 7, inplace=True)
df['Ventilation_System'].replace('3 Zone climate control', 8, inplace=True)
df['Ventilation_System'].replace('Fully automatic climate control, 3 Zone climate control', 9, inplace=True)
df['Ventilation_System'].replace('4 Zone climate control', 10, inplace=True)
df['Ventilation_System'].replace('Fully automatic climate control, 4 Zone climate control', 11, inplace=True)
```

```python
'''

Audiosystem
'No' < 'CD/MP3 Player' < 'CD Player with USB Only' < 'CD Player with Aux-in' < 'USB & Aux-in'< 
'CD Player with USB & Aux-in' < 'DVD Player with USB & Aux-in' < 'CD/MP3/DVD Player with USB & Aux-in'
  

'''

df["Audiosystem"].replace('No', 0, inplace=True)
df["Audiosystem"].replace('CD/MP3 Player', 1, inplace=True)
df["Audiosystem"].replace('CD Player with USB Only', 2, inplace=True)
df["Audiosystem"].replace('CD Player with Aux-in', 3, inplace=True)
df["Audiosystem"].replace('USB & Aux-in', 4, inplace=True)
df["Audiosystem"].replace('CD Player with USB & Aux-in', 5, inplace=True)
df["Audiosystem"].replace('DVD Player with USB & Aux-in', 6, inplace=True)
df["Audiosystem"].replace('CD/MP3/DVD Player with USB & Aux-in', 7, inplace=True)
```

```python
essentials = ['Make', 'Model', 'Variant', 'Displacement', 'Cylinders', 'Fuel_Tank_Capacity', 'ARAI_Certified_Mileage',
              'Power', 'Kerb_Weight']
```

```python
df['Seat_Height_Adjustment'] = df['Seat_Height_Adjustment']/df['Seat_Height_Adjustment'].max()
```

```python
df['Length'] = df['Length']/df['Length'].max()
```

```python
df['Width'] = df['Width']/df['Width'].max()
```

```python
df['Seating_Capacity'] = df['Seating_Capacity']/df['Seating_Capacity'].max()
```

```python
df['Boot_Space'] = df['Boot_Space']/df['Boot_Space'].max()
```

```python
comfort_features = ['Length', 'Width', 'Power_Seats', 'Power_Windows', 'Seats_Material', 'Audiosystem', 
                    'Boot_Space', 'Bluetooth', 'Ventilation_System', 'Parking_Assistance', 'Seat_Height_Adjustment',
                    'Seating_Capacity']
```

```python
security_features = ['Central_Locking', 'Child_Safety_Locks', 'Handbrake', 'Engine_Immobilizer', 
                     'ABS_(Anti-lock_Braking_System)', 'Door_Ajar_Warning', 'EBD_(Electronic_Brake-force_Distribution)',
                     'Fasten_Seat_Belt_Warning', 'EBA_(Electronic_Brake_Assist)', 'Distance_to_Empty']
```

```python
weights_comfort = [1,1,2,1.5,1,2,2.5,1,3,3,2,3]
```

```python
weights_security = [1.5,1.5,1,1.5,2,1.5,2,1.5,2,1]
```

```python
dfs = df[essentials]
```

```python
comfort = df['Length']

n = len(weights_comfort)
for i in range(1,n):
    comfort += weights_comfort[i]*df[comfort_features[i]]
```

```python
dfs['Comfort'] = comfort
```

```python
safety = df['Central_Locking']

m = len(weights_security)
for i in range(1,m):
    safety += weights_security[i]*df[security_features[i]]
```

```python
dfs['Safety'] = safety
```

```python
dfs['Price'] = df['Ex-Showroom_Price']
```

```python
dfs.rename(columns={'ARAI_Certified_Mileage' :'Mileage'}, inplace=True)
```
